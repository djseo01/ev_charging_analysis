#### 지도에 표시

```python
import folium
from folium.plugins import MarkerCluster
import pandas as pd

# CSV 파일 경로
csv_file_path = '인천_연수구_전기차 충전소.csv'

# CSV 파일 읽기
data = pd.read_csv(csv_file_path)

# NaN 값을 가진 행 제거
data = data.dropna(subset=['위도', '경도'])

# 지도의 중심을 설정할 위도와 경도 지정
map_center = [data['위도'].mean(), data['경도'].mean()]

# folium 지도 객체 생성
my_map = folium.Map(location=map_center, zoom_start=12)

# MarkerCluster 객체 생성
marker_cluster = MarkerCluster().add_to(my_map)

# CSV 데이터를 기반으로 지도에 마커 추가
for index, row in data.iterrows():
    icon = folium.Icon(color='red', prefix='fa', icon='bolt') 
    folium.Marker(
        location=[row['위도'], row['경도']],
        popup=row['충전소명'],  # 마커 클릭 시 나타날 팝업 텍스트
        icon=icon  # 마커 색상 지정
    ).add_to(marker_cluster)  # MarkerCluster에 마커 추가

my_map
```
