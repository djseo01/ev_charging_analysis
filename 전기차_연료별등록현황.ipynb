```python
import pandas as pd
import matplotlib.pyplot as plt

# 주어진 데이터
gasoline = df.loc[7, 'Unnamed: 20']
diesel = df.loc[24, 'Unnamed: 20']
lpg = df.loc[41, 'Unnamed: 20']
electronic = df.loc[75, 'Unnamed: 20']

# 데이터 구성
data = [electronic, lpg, diesel, gasoline]
labels = ['electronic', 'lpg', 'diesel', 'gasoline']

# 콤마 제거하고 최댓값을 기준으로 스케일링
max_value = max(float(val.replace(',', '')) for val in data)
scaled_data = [float(val.replace(',', '')) / max_value for val in data]

# 바 그래프 그리기
plt.figure(figsize=(10, 4))
plt.bar(labels, scaled_data, color='b')
plt.xlabel('Fuel Type')
plt.ylabel('Scaled Values')
plt.title('Values by Fuel Type (Scaled)')
plt.show()
```
